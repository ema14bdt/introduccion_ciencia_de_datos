**Introduccin:**

El dataset que estamos utilizando contiene informacin sobre diferentes circuitos de carreras, como los utilizados en la Frmula 1 y otros eventos de carreras de motor en todo el mundo. Cada fila del dataset representa un circuito especfico y proporciona detalles sobre su ubicacin geogrfica, pas, altitud y enlace a una fuente de informacin adicional, generalmente una pgina de Wikipedia. **Diccionario de Datos:**

A continuacin, se presenta un diccionario de datos que define las variables presentes en el dataset:

- `circuitId`: Un identificador nico para cada circuito. Este es un valor numrico.
- `circuitRef`: Una referencia o abreviatura del nombre del circuito. Es una cadena de caracteres.
- `name`: El nombre completo del circuito. Es una cadena de caracteres.
- `location`: La ubicacin geogrfica del circuito, que incluye la ciudad o regin donde se encuentra. Es una cadena de caracteres.
- `country`: El pas en el que se encuentra el circuito. Es una cadena de caracteres.
- `lat`: La latitud geogrfica del circuito. Es un valor numrico que representa la latitud en grados decimales.
- `lng`: La longitud geogrfica del circuito. Es un valor numrico que representa la longitud en grados decimales.
- `alt`: La altitud del circuito. Es un valor numrico que representa la altitud en metros sobre el nivel del mar.
- `url`: Un enlace a la pgina de Wikipedia u otra fuente que contiene informacin adicional sobre el circuito. Es una cadena de caracteres que representa la URL.

**Ejercicio 3**

Identificar el tipo de variable,decide justificando su respuesta.

- `circuitId`: Tipo de variable numrica. Este es un identificador nico para cada circuito y se representa mediante valores numricos enteros. No se utiliza para realizar clculos matemticos, por lo que es una variable categrica ordinal.
- `circuitRef`: Tipo de variable de texto (categrica nominal). Esta columna contiene referencias o abreviaturas del nombre del circuito. No se pueden realizar operaciones matemticas en estas cadenas, y cada valor es simplemente una etiqueta para identificar el circuito.
- `name`: Tipo de variable de texto (categrica nominal). Similar a la columna anterior, esta columna contiene el nombre completo del circuito y es de naturaleza puramente descriptiva.
- `location`: Tipo de variable de texto (categrica nominal). Al igual que las dos columnas anteriores, esta columna contiene informacin descriptiva sobre la ubicacin geogrfica del circuito.
- `country`: Tipo de variable de texto (categrica nominal). Esta columna representa el pas en el que se encuentra el circuito y es categrica, ya que no se pueden realizar operaciones matemticas en los nombres de los pases.
- `lat`: Tipo de variable numrica. La columna lat representa la latitud geogrfica de los circuitos y se expresa en grados decimales. Es una variable numrica porque se pueden realizar clculos matemticos con estos valores, como calcular la distancia entre circuitos basndose en su latitud.
- `lng`: Tipo de variable numrica. Similar a la columna lat, la columna lng representa la longitud geogrfica de los circuitos y se expresa en grados decimales. Tambin es una variable numrica porque se pueden realizar clculos con estos valores.
- `alt`: Tipo de variable numrica. La columna alt representa la altitud de los circuitos y se expresa en metros sobre el nivel del mar. Es una variable numrica porque se pueden realizar clculos matemticos con estos valores, como calcular la altitud promedio de los circuitos.
- `url`: Tipo de variable de texto (categrica nominal). La columna url contiene enlaces a pginas web que proporcionan informacin adicional sobre cada circuito. Es una variable categrica nominal, ya que no se pueden realizar operaciones matemticas en las URL.

```python
# Crear un histograma de latitud (lat)
plt.figure(figsize=(8, 6))
plt.hist(df['lat'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Histograma de Latitud (Latitud Geogrfica)')
plt.xlabel('Latitud')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Crear un histograma de longitud (lng)
plt.figure(figsize=(8, 6))
plt.hist(df['lng'], bins=20, edgecolor='k', alpha=0.7, color='orange')
plt.title('Histograma de Longitud (Longitud Geogrfica)')
plt.xlabel('Longitud')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

**Conclusiones:**

Distribucin de Latitudes (lat): El histograma de latitudes muestra que la mayora de los circuitos se encuentran en latitudes cercanas a 0. Esto es consistente con la ubicacin de la mayora de los circuitos de Frmula 1 en diferentes partes del mundo, lo que incluye reas tanto al norte como al sur del ecuador.

Distribucin de Longitudes (lng): El histograma de longitudes indica que los circuitos estn distribuidos en un amplio rango de longitudes. Esto es esperado, ya que los circuitos de carreras se encuentran en diferentes husos horarios y continentes, lo que resulta en una variabilidad considerable en las longitudes geogrficas.

```python
import numpy as np

# Reemplaza los valores "\N" en la columna 'alt' por NaN
df['alt'] = df['alt'].replace('\\N', np.nan)
# Convierte la columna 'alt' a tipo numrico (float)
df['alt'] = pd.to_numeric(df['alt'], errors='coerce')
altura_promedio_por_pais = df.groupby('country')['alt'].mean().reset_index()

# Ordena los pases por altura promedio de manera descendente
altura_promedio_por_pais = altura_promedio_por_pais.sort_values(by='alt', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(altura_promedio_por_pais['country'], altura_promedio_por_pais['alt'], color='skyblue')
plt.xlabel('Pas')
plt.ylabel('Altura Promedio (metros)')
plt.title('Altura Promedio por Pas')
plt.xticks(rotation=90)  # Rotar etiquetas del eje x
plt.tight_layout()

plt.show()
```

**Ejercicio 6**

Realizar analisis de matriz corelacion y explicar que variable estan correlacionadas\

```python
import seaborn as sns

columnas_numericas = ['lat', 'lng']

matriz_correlacion = df[columnas_numericas].corr(numeric_only=True)

# Visualiza la matriz de correlacin utilizando un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlacin')
plt.show()
```

Correlacin Positiva: Si dos variables tienen una correlacin positiva, significa que cuando una variable aumenta, la otra tiende a aumentar tambin. En la matriz de correlacin, vers valores cercanos a 1. Por ejemplo, si lat y lng tienen una correlacin positiva alta, esto podra indicar que a medida que te desplazas hacia el norte en latitud, es ms probable que te desplaces hacia el este en longitud.

Correlacin Negativa: Si dos variables tienen una correlacin negativa, significa que cuando una variable aumenta, la otra tiende a disminuir. Vers valores cercanos a -1 en la matriz de correlacin. Por ejemplo, si lat y alt tienen una correlacin negativa, esto podra indicar que a medida que te desplazas hacia el norte en latitud, es menos probable que te encuentres a una altitud elevada.
