**Introducción:**

El dataset que estamos utilizando contiene información sobre diferentes circuitos de carreras, como los utilizados en la Fórmula 1 y otros eventos de carreras de motor en todo el mundo. Cada fila del dataset representa un circuito específico y proporciona detalles sobre su ubicación geográfica, país, altitud y enlace a una fuente de información adicional, generalmente una página de Wikipedia. Estos datos son valiosos para realizar análisis geográficos y estadísticos sobre circuitos de carreras, así como para explorar las características de diferentes ubicaciones donde se llevan a cabo eventos automovilísticos.

**Diccionario de Datos:**

A continuación, se presenta un diccionario de datos que define las variables presentes en el dataset:

- `circuitId`: Un identificador único para cada circuito. Este es un valor numérico.
- `circuitRef`: Una referencia o abreviatura del nombre del circuito. Es una cadena de caracteres.
- `name`: El nombre completo del circuito. Es una cadena de caracteres.
- `location`: La ubicación geográfica del circuito, que incluye la ciudad o región donde se encuentra. Es una cadena de caracteres.
- `country`: El país en el que se encuentra el circuito. Es una cadena de caracteres.
- `lat`: La latitud geográfica del circuito. Es un valor numérico que representa la latitud en grados decimales.
- `lng`: La longitud geográfica del circuito. Es un valor numérico que representa la longitud en grados decimales.
- `alt`: La altitud del circuito. Es un valor numérico que representa la altitud en metros sobre el nivel del mar.
- `url`: Un enlace a la página de Wikipedia u otra fuente que contiene información adicional sobre el circuito. Es una cadena de caracteres que representa la URL.

**Ejercicio 3**

Identificar el tipo de variable,decide justificando su respuesta.

- `circuitId`: Tipo de variable numérica. Este es un identificador único para cada circuito y se representa mediante valores numéricos enteros. No se utiliza para realizar cálculos matemáticos, por lo que es una variable categórica ordinal.
- `circuitRef`: Tipo de variable de texto (categórica nominal). Esta columna contiene referencias o abreviaturas del nombre del circuito. No se pueden realizar operaciones matemáticas en estas cadenas, y cada valor es simplemente una etiqueta para identificar el circuito.
- `name`: Tipo de variable de texto (categórica nominal). Similar a la columna anterior, esta columna contiene el nombre completo del circuito y es de naturaleza puramente descriptiva.
- `location`: Tipo de variable de texto (categórica nominal). Al igual que las dos columnas anteriores, esta columna contiene información descriptiva sobre la ubicación geográfica del circuito.
- `country`: Tipo de variable de texto (categórica nominal). Esta columna representa el país en el que se encuentra el circuito y es categórica, ya que no se pueden realizar operaciones matemáticas en los nombres de los países.
- `lat`: Tipo de variable numérica. La columna lat representa la latitud geográfica de los circuitos y se expresa en grados decimales. Es una variable numérica porque se pueden realizar cálculos matemáticos con estos valores, como calcular la distancia entre circuitos basándose en su latitud.
- `lng`: Tipo de variable numérica. Similar a la columna lat, la columna lng representa la longitud geográfica de los circuitos y se expresa en grados decimales. También es una variable numérica porque se pueden realizar cálculos con estos valores.
- `alt`: Tipo de variable numérica. La columna alt representa la altitud de los circuitos y se expresa en metros sobre el nivel del mar. Es una variable numérica porque se pueden realizar cálculos matemáticos con estos valores, como calcular la altitud promedio de los circuitos.
- `url`: Tipo de variable de texto (categórica nominal). La columna url contiene enlaces a páginas web que proporcionan información adicional sobre cada circuito. Es una variable categórica nominal, ya que no se pueden realizar operaciones matemáticas en las URL.

Realizar un analisis univariado y en base a esos graficos,sacar conclusiones.

```python
# Crear un histograma de latitud (lat)
plt.figure(figsize=(8, 6))
plt.hist(df['lat'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Histograma de Latitud (Latitud Geográfica)')
plt.xlabel('Latitud')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Crear un histograma de longitud (lng)
plt.figure(figsize=(8, 6))
plt.hist(df['lng'], bins=20, edgecolor='k', alpha=0.7, color='orange')
plt.title('Histograma de Longitud (Longitud Geográfica)')
plt.xlabel('Longitud')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

**Conclusiones:**

Distribución de Latitudes (lat): El histograma de latitudes muestra que la mayoría de los circuitos se encuentran en latitudes cercanas a 0. Esto es consistente con la ubicación de la mayoría de los circuitos de Fórmula 1 en diferentes partes del mundo, lo que incluye áreas tanto al norte como al sur del ecuador.

Distribución de Longitudes (lng): El histograma de longitudes indica que los circuitos están distribuidos en un amplio rango de longitudes. Esto es esperado, ya que los circuitos de carreras se encuentran en diferentes husos horarios y continentes, lo que resulta en una variabilidad considerable en las longitudes geográficas.

```python
import numpy as np

# Reemplaza los valores "\N" en la columna 'alt' por NaN
df['alt'] = df['alt'].replace('\\N', np.nan)
# Convierte la columna 'alt' a tipo numérico (float)
df['alt'] = pd.to_numeric(df['alt'], errors='coerce')
altura_promedio_por_pais = df.groupby('country')['alt'].mean().reset_index()

# Ordena los países por altura promedio de manera descendente
altura_promedio_por_pais = altura_promedio_por_pais.sort_values(by='alt', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(altura_promedio_por_pais['country'], altura_promedio_por_pais['alt'], color='skyblue')
plt.xlabel('País')
plt.ylabel('Altura Promedio (metros)')
plt.title('Altura Promedio por País')
plt.xticks(rotation=90)  # Rotar etiquetas del eje x
plt.tight_layout()

plt.show()
```

**Ejercicio 6**

Realizar analisis de matriz corelacion y explicar que variable estan correlacionadas\

```python
import seaborn as sns

columnas_numericas = ['lat', 'lng']

matriz_correlacion = df[columnas_numericas].corr(numeric_only=True)

# Visualiza la matriz de correlación utilizando un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

Correlación Positiva: Si dos variables tienen una correlación positiva, significa que cuando una variable aumenta, la otra tiende a aumentar también. En la matriz de correlación, verás valores cercanos a 1. Por ejemplo, si lat y lng tienen una correlación positiva alta, esto podría indicar que a medida que te desplazas hacia el norte en latitud, es más probable que te desplaces hacia el este en longitud.

Correlación Negativa: Si dos variables tienen una correlación negativa, significa que cuando una variable aumenta, la otra tiende a disminuir. Verás valores cercanos a -1 en la matriz de correlación. Por ejemplo, si lat y alt tienen una correlación negativa, esto podría indicar que a medida que te desplazas hacia el norte en latitud, es menos probable que te encuentres a una altitud elevada.
