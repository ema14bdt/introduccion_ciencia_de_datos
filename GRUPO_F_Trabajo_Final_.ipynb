```python
#1.IMPORTAR LIBRERIAS:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import datetime as dt
import requests
from google.colab import drive
from sklearn import preprocessing, linear_model,model_selection
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

#2.CARGAR Y EXPLORACION DEL DATASET:
# Ruta del archivo CSV descargado en Google Colab
ruta_csv = '/content/ruta_csv/MyDrive/Colab_Notebooks/registro_temperatura365-ordenado-y-separado.csv'
drive.mount('ruta_csv', force_remount=True)
df = pd.read_csv(ruta_csv, encoding='latin-1')
## EXPLORACION DE DATOS
## Vista de las primeras 5 filas
print(df.head(5))
print("\n\n")
```

```python
##3. PROCESO DE LIMPIEZA
data_frame = df.isnull().sum() #Verifica la cantidad de valores nulos en cada columna, los suma y lo guarda
data_frame = df.dropna() ## Eliminamos valores nulos
data_frame = df.round(2) # Redondeo de decimales a 2 unidades
data_frame.describe().round(2)
```

```python
## Dejamos en el dataFrame solo las filas de la ciudad elegida.

df.drop(df[(df['NOMBRE'] !=' COMODORO RIVADAVIA AERO')].index, inplace=True)

df
```

```python
# Calculamos la temperatura promedio por dia

df['TMAX'] = pd.to_numeric(df['TMAX'],errors='coerce')
df['TMIN'] = pd.to_numeric(df['TMIN'],errors='coerce')
df['TEMP_PROM'] = (df['TMAX'] + df['TMIN'])/2
df
```

```python
# Separamos los datos en variables independientes (X - tiempo) y dependientes (Y - Temperatura Promedio) - Seran usados en Regresion Lineal

X = df[['FECHA']]
y= df['TEMP_PROM']

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.5, random_state=0) # Cjto de entrenamiento y prueba para temperatura promedio

## Aplicamos Regresion Lineal

## Creamos el modelo de Regresion Lineal
modelo = LinearRegression()

# Entrenamos el modelo con los datos de entrenamiento
modelo_temperatura = modelo.fit(X_train, Y_train)

# Realizamos predicciones en los datos de prueba
Y_pred_e = modelo_temperatura.predict(X_test)


## Se va a hacer la correlacion entre el tiempo y la temperatura promedio
## En Regresion Lineal
x = np.array(df['FECHA'])
temperatura=modelo_temperatura.fit(x.reshape(-1,1), df['TEMP_PROM'])

# Hacemos la prediccion para los 3 medios de pagos
prediccion_temperatura = temperatura.predict(x.reshape(-1,1))

## GRAFICAMOS la regresion Lineal entre el tiempo y la temperatura promedio

plt.figure(figsize=(10, 6))
plt.scatter(df['FECHA'], df['TEMP_PROM'], color='blue', label='Temperatura', marker ='o',s=100)
plt.plot(x, prediccion_temperatura, color='green', marker='*',linewidth=2, label='Regresión Lineal')
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Temperatura')
plt.title('Comparacion de temperatura en el Tiempo')
plt.show()
print("\n\n\n")

```

```python
## 2: Agregamos las columnas Precio helado

precio_inicial = 10.0  # Precio base del helado
porcentaje_incremento = 0.3  # Porcentaje de incremento por grado Celsius

# Generamos precios de helado aleatorios en función de la temperatura
precios_helado = precio_inicial + porcentaje_incremento * (df['TEMP_PROM'])

# Añadimos la columna de precios de helado
df['PRECIO DEL HELADO'] = precios_helado
print(df['PRECIO DEL HELADO'])
```

```python
## 2.2: Agregamos la columna Compra helado

# Generamos la columna "COMPRA HELADO"
def decide_compra(precio_actual, precio_inicial, probabilidad_no_compra=0.8):
    if precio_actual > 1.4 * precio_inicial:
        return "NO" if np.random.rand() < probabilidad_no_compra else "SI"
    else:
        return "SI"

df['COMPRA HELADO'] = df.apply(lambda row: decide_compra(row['PRECIO DEL HELADO'], precio_inicial), axis=1)
print(df['COMPRA HELADO'])
print(df)
```

```python
## 3: Perceptrones

# Definimos pesos aleatorios para cada perceptrón
pesos_perceptron3 = np.random.rand(2)

# Función de activación (sigmoide)
def sigmoide(x):
    return 1 / (1 + np.exp(-x))

# Definir la función para evaluar la decisión de cada perceptrón
def evaluar_perceptron1(temperatura):
    return temperatura > 27

def evaluar_perceptron2(precio_helado):
    return precio_helado < precio_inicial * 1.4  # 40% de incremento

def evaluar_perceptron3(decision_perceptron1, decision_perceptron2, pesos):
    entrada = np.array([decision_perceptron1, decision_perceptron2])
    suma_ponderada = np.dot(entrada, pesos)
    activacion = sigmoide(suma_ponderada)
    return activacion
```

```python
##4: Implementacion de la red neuronal

temperaturas = df['TEMP_PROM']
precios_helado = df['PRECIO DEL HELADO']
compra_helado_real = df['COMPRA HELADO']

decisiones_red_neuronal = []

for i in range(len(df)):
    row = df.iloc[i]

    decision_perceptron1 = evaluar_perceptron1(row['TEMP_PROM'],)
    decision_perceptron2 = evaluar_perceptron2(row['PRECIO DEL HELADO'], )
    decision_perceptron3 = evaluar_perceptron3(decision_perceptron1, decision_perceptron2, pesos_perceptron3)

    if decision_perceptron3 > 0.5:
        decisiones_red_neuronal.append("SI")
    else:
        decisiones_red_neuronal.append("NO")

aciertos = 0

for i in range(len(decisiones_red_neuronal)):
    if decisiones_red_neuronal[i] == compra_helado_real.iloc[i]:
        aciertos += 1

proporcion_aciertos = aciertos / len(decisiones_red_neuronal) * 100
print(f"La red neuronal acierta en un {proporcion_aciertos} % de los casos.")
```
